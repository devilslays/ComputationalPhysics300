```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#this function handles the iteration of the complex function. For the mandelbrot set, the value of z
#starts from 0 and is constantly updated with each iteration
MAX_VALUE = 300
def mandel(c):
    z = 0
    n = 0
    c = complex(*c)
    while (abs(z) <= 2) and (n < MAX_VALUE):
        z = z**2 + c
        n += 1
    return n
```

```python
res = 500
#this sets the x min and max for the function
x_min, x_max = -2, 2
y_min, y_max = -2, 2
#this creates a pair of coordinates
x0, y0 = np.meshgrid(np.linspace(x_min, x_max, res), np.linspace(y_min, y_max, res))
#the np.c_ function combines two matrices by row
values = np.c_[x0.ravel(),y0.ravel()]
#this line of code relates the function for the iteration to a variable that
#will be used to show the plot of the mandelbrot set
mandelbrot = [mandel(c) for c in values]
mandelbrot = np.array(mandelbrot).reshape(500,500)
plt.imshow(mandelbrot)
plt.show()
```
