```python
from numpy import fft
import cmath
import numpy as np
import matplotlib.pyplot as plt
from numpy import loadtxt
```

```python
#load the txt file with the data
ytxt = np.loadtxt("circular.txt")
plt.plot(ytxt)
plt.savefig("uncompressed image")
plt.show()
```

```python
c = fft.fftn(ytxt)
print(len(c))
plt.plot(np.abs(c))
plt.show()
```

```python
compressed_image = fft.fftn(c).real
plt.imshow(compressed_image)
plt.savefig("compressed image")
plt.show()
```
