```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np
```

```python
#main curve represented by a cosine curve
f = lambda x,a,k: a*np.cos(k*x)
x_min, x_max = 0, 10
npoints = 75
order = 20
order2 = 10
a, k = 1.2, 1.2

x = np.linspace(x_min, x_max, npoints)
y = f(x, a, k) + np.random.rand(npoints)

#different fit functions
fit = np.polyfit(x,y,order,full=True)
fit2 = np.polyfit(x,y,order2,full=True)
y_p = np.polyval(fit[0],x)
y_p2 = np.polyval(fit2[0],x)

plt.plot(x,f(a,k,x),label='actual function')
plt.plot(x,y,'o',label ='noisy data')
plt.plot(x,y_p,label='polyfit_'+str(order))
plt.plot(x,y_p2,label='polyfit2_'+str(order2))

plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.legend(fontsize=12)
plt.show()

#it looks like a polynomial fit with a high order would be the best fit for this function
```
