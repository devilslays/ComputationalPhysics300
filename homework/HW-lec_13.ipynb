```python
#this is the function which repeats the dice rolling N amount of times
#then counts how many times double six is rolled
def diceroll(N):
    count = 0
    for i in range(N):
        dice1=np.random.randint(1,7)
        dice2=np.random.randint(1,7)
        if (dice1 == 6) and (dice2 == 6):
            count +=1 
    print('you rolled double six',count,'times!')
    #this is the probability of rolling double six
    probability = count/N
    print('the probability of rolling a six is',probability)
```

```python
diceroll(1000000)
```

```python
#this is part c of the assignment, which is to find the magic series
N = 100
mseries=[]
for i in range(N):
    x=np.random.randint(-5,5)
    mseries.append(x)
plt.plot(mseries,'o')
plt.show()
plt.hist(mseries)
plt.show()
```
