```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
from random import random
from math import pi
```

```python
#definition of the function
def f(x):
    return 1/x*(1/(np.exp(x)+1))
                
def w(x):
    return 1/x

def p(x):
    return x
```

```python
#plotting the graph of the two functions
x = np.linspace(0,1,100)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, w(x), label='w(x)')
plt.legend()
plt.show()
```

```python
print('from scipy:', quad(f,0,1)[0])
```

```python
from random import random
f = lambda x: 1/x*(1/(np.exp(x)+1))
w = lambda x: 1/x
p = lambda x: x
```

```python
def IMC(N):
    const = quad(w,0,1)[0]
    I = 0
    #the for loop that handles the random number as well as calculatig
    #I for the IMC.
    for i in range(0,N):
        x = random()
        y = p(x)
        I += f(y)/w(y)
    return I/N*const

print(IMC(1000000))
```
