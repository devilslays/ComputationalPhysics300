```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
from random import random
import statistics
from math import sqrt
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, aspect='equal')

x1 = np.linspace(0,2,100)
y1 = np.sqrt(1-(x1-1)**2)
ax1.plot(x1,y1,label = 'y1')

y2 = 2-np.sqrt(4-x1**2)
ax1.plot(x1,y2,label = 'y2')

ax1.legend()
ax1.set_xlim(0,1.75)
ax1.set_ylim(0,1.25)
ax1.fill_between(x1, y1, y2, where=y1>y2, facecolor='green')
#creates a rectangle in order to find the area to be used in the
#equation for I
rect = patches.Rectangle((0,0),1.6,1,linewidth=1, edgecolor='r',facecolor='none')
ax1.add_patch(rect)
plt.show()
plt.savefig('MC1.png')
```

```python
#define the functions of the curves to plug into the MC function
def f(x):
    return np.sqrt(1-(x-1)**2)

def g(x):
    return 2-np.sqrt(4-x**2)

#this defines the monte carlo function and uses a nest for and if loop
#in order to count how many random points are in the shaded region
def MC(N):
    count = 0
    for i in range(N):
        x = 1.6*random()
        y = random()
        if ((y<f(x)) and (y>g(x))):
            count += 1
    I = 1.6*count/N
    return I
```

```python
print(MC(100))
print(MC(1000))
print(MC(10000))
print(MC(100000))
```

```python
list = [MC(100), MC(1000), MC(10000), MC(100000)]
std = statistics.stdev(list)
print ("the standard deviation is", std)
```

```python
list2 = [MC(100), MC(1000), MC(10000), MC(100000)]
var = statistics.variance(list2)
print("the variance is", var)
```
