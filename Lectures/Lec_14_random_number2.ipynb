## 14.1 Brownian Motion
Brownian motion is the motion of a particle, such as a smoke or dust paticle, in a gas, as it is buffered by random collisions
with gas molecules. Make a simple computer simulation of such a particle in two dimensions as follows.

- 1, __Initilization__. The particle is confined to a square grid or lattice L\*L squares on a side, so that its poistion can be represented by two integers i,j = 0 ... L-1. It starts in the middle of the grid. 
- 2, __Random Walk__. On each step of the simulation, choose a random direction - up, down, left, or right - and move the particle one step in that direction
- 3, __Boundary condition__. If the particle goes outside the limits of the lattice, reject it, and choose a new random direction.

Write a program to perform a 1000 steps of this process on a lattice with $L$ = 101.


````python
# Brownian Motion
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=[18, 12])

# initialization
L1 = 101
L2 = 101
N = 1000

position = np.zeros([N,2])
position[0,0] = int(L1/2)
position[0,1] = int(L2/2)

direction = np.array([[0,1],[0,-1],[1,0],[-1,0]]) #up, down, right, left

#np.random.seed(2) 

# random walk
for i in range(1,N):    
    while True:
        move = np.random.randint(len(direction))
        position_next = position[i-1,:] + direction[move]
        #boundary condition
        if 0<position_next[0]+1<L1+2 and 1<position_next[1]+1<L2+2:
            break
    
    #update the positions for step i.
    position[i,0] = position_next[0]
    position[i,1] = position_next[1]
    #position[i,:] = position_next

# plot the trajectory
ax = plt.axes()
ax.set_facecolor('silver')
plt.xlim(0,L1)
plt.ylim(0,L2)
plt.title('Brownian motion')
plt.scatter(position[:,0], position[:,1], c=range(N), s=60, alpha=0.8)
plt.text(position[0,0], position[0,1], 'start', fontsize=24, color='w')
plt.text(position[-1,0], position[-1,1], 'end', fontsize=24)

cbar = plt.colorbar()
cbar.set_label('Itertion number')
plt.show()
````

## 14.2 Non-uniform Random Numbers

All the python random number functions we have decribed generate uniformly distributed random numbers - all values that can 
be returned by these functions occur with equal probability. In physics, however, there are many processes that happen
with nonuniform probability. 

### 14.2.1 Decay of an isotope
The radioisotope $^{201}$Tl decays to $^{208}$Pb with a half-life of 3.053 minutes. Suppose we start with a sample of 1000 thallium atoms. Let us simulate the decay of these atoms over time, mimicking the randomness of that decay using random numbers.

On average we know that the number N of atoms in our sample will fall off exponentially over time according to the standard equation of radioactive decay:
$$ N(t) = N(0) 2^{-t/\tau}$$

where $\tau$ is the half-life. Then the fraction of atoms remaining after time $t$ is 

$$N(t)/N(0) = 2^{-t/\tau}$$

Which means the probability $p(t)$ that any particular single atom has decayed is:

$$p(t)=1-2^{-t/\tau}$$

Simulate the decay process as a function of time.


````python
## Decay process
import numpy as np
import matplotlib.pyplot as plt


# Initial parameters
tmax = 1000
N_Tl = 1000
N_Pb = 0
tau = 3.053*60 #seconds
dt = 1 #seconds
p = 1 - 2**(-dt/tau)

t_points = np.arange(0.0, tmax, dt)
Tl = []
Pb = []

# N_time * N_atoms
for t in range(tmax): #t_points:
    Tl.append(N_Tl)
    Pb.append(N_Pb)
    
    # Calculate the decayed atoms
    decay = 0
    for i in range(N_Tl):
        if np.random.random()<p:
            decay += 1
    N_Tl -= decay
    N_Pb += decay

plt.plot(t_points, Tl, label='Tl')
plt.plot(t_points, Pb, label='Pb')
plt.xlabel("Time")
plt.ylabel("Number of atoms")
plt.legend()
plt.show()
````

As shown in the above, the probability that a atom decays in time $t$ is $1-2^{-t/\tau}$, and hence the probability that it decays in a time interval $dt$ is
$$1-2^{-dt/\tau} = 1-\exp(-\frac{dt}{\tau}\text{ln}2) = \frac{\text{ln}2}{\tau}dt$$

Then we calculate the total probability of decaying an atom between $t$ and $t+dt$. In order to undergo such a decay,
an atom must survive without decay until time $t$, which happens with $p=2^{-t/\tau}$, and then decay in dt interval:

$$P = P(t)P(dt) = 2^{-t/\tau}\frac{\text{ln}2}{\tau}dt$$

To make it easier, we set $\mu=\text{ln}2/\tau$.

$$p(t)=\mu e^{-\mu t}$$

If you do integration from 0 to infinity:

$$\mu \int_0^{\infty} e^{-\mu t} t = 1$$

**This is an example of non-uniform probability distribution. It indidates that the decay times `t` are distributed in proportion to $2^{-t/\tau}$.**


````python
# To generate the exponential distribution from uniform distribution
import numpy as np
import matplotlib.pyplot as plt

tau = 3.053*60
t = np.linspace(0,1000,100)

y = (2**(-t/tau)*np.log(2)/tau)

Tl = []
Pb = []
NTl = 1000
NPb = 0

plt.plot(t,y)
plt.xlim(0,1000)
plt.show()
````

Since we known the distribution of `t`, a more efficient way to calculate the decay of a sample is to generate `N` random numbers from this non-uniform probability distribution, to represent the time at which each of the atom decays. Then creating curves as a function of time, is simply a matter of counting how many of the atoms decay before any given time.

### Generate the non-uniform random numbers

It is possible to generate non-uniform random numbers from a wide range of different distributions, including the exponential distribution, using the previous method. If you already have a source of uniform random numbers, one could turn them into nonuniform ones using some `transformation methods`.

*Suppose you have a source of random floating numbers `z` from a distribution with a probability of `q(z)`. If you have a function `x(z)`. Then when `z` is one of our random numbers, `x(z)` is also a random number, but it will have different distribution according to `x(z)`.*

$$p(x)dx = q(z)dz$$
$$\int_{-\infty}^{x(z)} p(x')dx' = \int_0^z dz'=z$$

### Solution of the Decay Time
For the distribution of decay time, the cumulative probability ($F$) is
$$\mu \int_0^{x(z)} e^{-\mu x'} dx' = 1-e^{-\mu x} = z = F$$

Therefore, we can get the inverse of $F$:

$$x = -\frac{1}{\mu}\text{ln}(1-z)$$

So all we need is to generate uniform random numbers `t` in the interval from zero to one and feed them into this equation to get exponentially distributed `x` values.

````python
# Code to test the random numbers are following the desired distribution

N = 1000
mu = np.log(2)/tau
x = []
for i in range(N):
    Tl.append(NTl)
    Pb.append(NPb)
    
    z0 = np.random.random() #uniform random number
    x.append(-np.log(1-z0)/mu) #nonuniform number that follows a distribution
    
plt.hist(x,100) #, normed=1)
plt.xlim(0,1000)
plt.show()
````
